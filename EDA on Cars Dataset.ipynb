```python
data = [7,6,5,10]
plt.title('Distribution visualization using Pie Diagram')
plt.pie(data, explode=[0.1, 0.1, 0, 0], labels = ['t1', 't2', 't3', 't4'], autopct = "%1.1f%%", shadow = True, colors = ['yellow', 'green', 'cyan', 'red'])
plt.show()
```
